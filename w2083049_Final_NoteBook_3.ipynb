## Load the Dataset

```python
import pandas as pd
print("===========================================================")
print("Load the Classification dataset (from Notebook 1)")
print("===========================================================")

df = pd.read_csv('/content/classification_dataset.csv')

# Display basic info
df.head()
```

```python
df.info()
```

```python
df.describe()
```

##Selecting Feature

```python
# Define target and features
X = df.drop("Mortality_Status", axis=1)
y = df["Mortality_Status"]

print("==== x data ====")
print(X.head())
print("\n==== y data ====\n")
print(y.head())
```

### Divide data set training and testing

```python
from sklearn.model_selection import train_test_split

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

### Data Standardization

```python
from sklearn.preprocessing import StandardScaler

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Build Base Learners & Ensemble

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

# Base learners
lr = LogisticRegression(max_iter=1000)
nb = GaussianNB()

# Voting Ensemble
ensemble = VotingClassifier(estimators=[('lr', lr),('nb', nb)], voting='soft')

ensemble.fit(X_train_scaled, y_train)
y_pred_ensemble = ensemble.predict(X_test_scaled)
```

## Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

print(confusion_matrix(y_test, y_pred_ensemble))
print(classification_report(y_test, y_pred_ensemble))

# AUC-ROC
y_proba = ensemble.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_proba)
auc = roc_auc_score(y_test, y_proba)

plt.plot(fpr, tpr, label=f'Ensemble AUC = {auc:.2f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Ensemble ROC Curve')
plt.legend()
plt.show()
```

# Part B: Decision Tree Regression for Survival_Months

## Load Regression Dataset

```python
import pandas as pd
print("===========================================================")
print("Load the Regression dataset (from Notebook 1)")
print("===========================================================")

data_frame = pd.read_csv('/content/classification_dataset.csv')

# Display basic info
data_frame.head()
```

### Selecting Feature

```python
# Define target and features
X = data_frame.drop("Survival_Months", axis=1)
y = data_frame["Survival_Months"]

print("==== x data ====")
print(X.head())
print("\n==== y data ====\n")
print(y.head())
```

### Divide data set training and testing

```python
from sklearn.model_selection import train_test_split

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Regression Model Evaluation Metrics: MAE, MSE, RMSE, R²

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_regression(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    return mae, mse, rmse, r2
```

# Train and Evaluate Classifiers

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor(random_state=42)
tree.fit(X_train_scaled, y_train)
tree_mae, tree_mse, tree_rmse, tree_r2 = evaluate_regression(tree, X_test_scaled, y_test)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42, n_estimators=100)
rf.fit(X_train_scaled, y_train)
rf_mae, rf_mse, rf_rmse, rf_r2 = evaluate_regression(rf, X_test_scaled, y_test)
```

### Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(random_state=42)
gb.fit(X_train_scaled, y_train)
gb_mae, gb_mse, gb_rmse, gb_r2 = evaluate_regression(gb, X_test_scaled, y_test)
```

```python
results = pd.DataFrame({
    'Model': ['Decision Tree', 'Random Forest', 'Gradient Boosting'],
    'MAE': [tree_mae, rf_mae, gb_mae],
    'MSE': [tree_mse, rf_mse, gb_mse],
    'RMSE': [tree_rmse, rf_rmse, gb_rmse],
    'R² Score': [tree_r2, rf_r2, gb_r2]
})

results.sort_values(by='RMSE')
```
