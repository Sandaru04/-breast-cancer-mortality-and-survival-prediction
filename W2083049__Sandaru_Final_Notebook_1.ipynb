# Data Drop , fill in missing values and Encode Categorical Variables

```python
# Drop Patient ID, Month of Birth and Occupation (irrelevant or with many missing values)
data_frame = data_frame.drop(['Patient_ID', 'Month_of_Birth', 'Occupation'], axis=1)
print("Data Drop sussecfully")
```

```python
# Fill 'Age' missing values with median
data_frame['Age'] = data_frame['Age'].fillna(data_frame['Age'].median())

# Fill 'Sex' missing values with mode
data_frame['Sex'] = data_frame['Sex'].fillna(data_frame['Sex'].mode()[0])

# Fill 'Tumor_Size' missing values with median
data_frame['Tumor_Size'] = data_frame['Tumor_Size'].fillna(data_frame['Tumor_Size'].median())

# Fill 'Regional_Node_Examined' with median
data_frame['Regional_Node_Examined'] = data_frame['Regional_Node_Examined'].fillna(data_frame['Regional_Node_Examined'].median())
```

```python
# Standardize text
data_frame['Mortality_Status'] = data_frame['Mortality_Status'].str.strip().str.lower()
```

```python
# Importing plotly.express for creating visualizations
import plotly.express as px

# Create a histogram for the Age column
Mortality_Status_fig = px.bar(data_frame, x='Mortality_Status')
Mortality_Status_fig.show()
```

```python
# Encode categorical variables
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
mortality_encoder = preprocessing.LabelEncoder()
data_frame['Sex'] = label_encoder.fit_transform(data_frame['Sex'])
data_frame['Estrogen_Status'] = label_encoder.fit_transform(data_frame['Estrogen_Status'])
data_frame['Progesterone_Status'] = label_encoder.fit_transform(data_frame['Progesterone_Status'])
data_frame['Mortality_Status'] = mortality_encoder.fit_transform(data_frame['Mortality_Status'])
data_frame['A_Stage'] = label_encoder.fit_transform(data_frame['A_Stage'])
data_frame['T_Stage'] = label_encoder.fit_transform(data_frame['T_Stage'])
data_frame['N_Stage'] = label_encoder.fit_transform(data_frame['N_Stage'])
data_frame['Differentiated'] = label_encoder.fit_transform(data_frame['Differentiated'])
data_frame['6th_Stage'] = label_encoder.fit_transform(data_frame['6th_Stage'])
data_frame.head()
```

## Finding Outliers Age  Column


```python
# Importing plotly.express for creating visualizations
import plotly.express as px

# Create a histogram for the Age column
age_fig = px.histogram(data_frame, x='Age')
age_fig.show()

# Create a box plot for the Age column
age_fig = px.box(data_frame, x='Age')
age_fig.show()

# Create a scatter plot for the Age column
age_fig = px.scatter(data_frame, x='Age')
age_fig.show()
```

```python
def find_outliers_IQR(df):
  q1=df.quantile(0.25)
  q3=df.quantile(0.75)
  IQR=q3-q1
  outliers = df[((df<(q1-1.5*IQR))|(df>(q3+1.5*IQR)))]
  return outliers

outliers = find_outliers_IQR(data_frame['Age'])
print("number of outliers: "+ str(len(outliers)))
outliers
```
